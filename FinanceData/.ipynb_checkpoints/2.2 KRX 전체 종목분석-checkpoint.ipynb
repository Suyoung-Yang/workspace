## Boolean Indexing
    * 결과값이 true,false 형태로 반환되기 때문에 boolean indexing
    * 특정 조건식을 만족하는 데이터를 서브셋으로 가져오기

```python
# df["Name"] == "카카오" # 시리즈 형태로 나오고, 원하는 결과가 나오지 않는다.
df[df["Name"] == "카카오"] # 데이터 프레임 형태를 씌워서 원하는 정보를 얻는다.
```

```python
# and => &
# or => |
```

```python
# 서울특별시에 소재지를 두고 있는 KOSPI 종목의 종목명과 심볼을 가져옵니다.
# df[df["Market"] == "KOSPI"]
df[(df["Region"] == "서울특별시") & (df["Market"] == "KOSPI")] # 연산자 우선순위를 위해 ()를 감싸준다.

# loc를 통해 원하는 열만 가져올 수 있음.
df.loc[(df["Region"] == "서울특별시") & (df["Market"] == "KOSPI"), ["Name","Symbol"]]

```

# 파생변수 만들기

```python
#pd.to_datetime을 통해 데이터 타입 변환하기
df["ListingDate"] # dtype이 object이므로, 날짜로 변환이 필요
pd.to_datetime(df["ListingDate"]) # 데이터 타입 변환됨
```

```python
# ListingYear 파생변수 만들기 
df["ListingDate"] # 아직 변하지 않았음. 왜냐하면, 해당 데이터에 할당하지 않았기 때문임.
df["ListingDate"] = pd.to_datetime(df["ListingDate"]).copy()
```

```python
# Listing Year 파생변수 만들기
df["ListingDate"].dt.month # dt access를 통해 year, month, day를 가져올 수 있다.
df["ListingYear"] = df["ListingDate"].dt.year.copy()
```

- KOSPI (Korea Composite Stock Price Index)
- 우리나라의 종합주가지수(KOSPI)는 증권거래소가 1964.1.4일을 기준시점으로 다우존스식 주가평균을 지수화한 수정주가 평균지수를 산출, 발표하기 시작하였다. 다우존스식 주가지수는 주가지수를 구성하는 상장종목 중 일부 우량주만을 선정하여 산출하는 방식이다. 시가총액식 주가지수는 일정시점의 시가총액과 현재시점의 시가총액을 대비, 현재의 주가수준을 판단하는 방식이다. 즉, 지난 1980년 1월 4일 기준 상장종목 전체의 시가총액을 100으로 보고 현재 상장종목들의 시가총액이 어느 수준에 놓여 있는지를 보여주는 시스템이다.
종합주가지수 = (비교시점의 시가총액 ÷ 기준시점의 시가총액) × 100
출처 : [네이버 지식백과] 종합주가지수 (시사상식사전, pmg 지식엔진연구소) https://terms.naver.com/entry.nhn?docId=72147&cid=43667&categoryId=43667

- KOSDAQ (Korea Securities Dealers Automated Quotation)

```python
# groupby([인덱스 컬럼 리스트])[연산에 사용할 컬럼].연산()
df_year_market = df.groupby(["ListingYear","Market"])["Symbol"].count().reset_index(name="count")
```

```python
# pivot_table로 같은 결과가 나오게 연산하기
# 1. pd. 이용하는 방법. 2. df. 이용하는 방법
# pd.pivot_table(data=df,index="ListingYear",values="Symbol",aggfunc="count")
df.pivot_table(index=["ListingYear","Market"],values="Symbol",aggfunc="count").reset_index()
```

```python
# df_year_market.rename(columns={""})
```

```python
plt.figure(figsize=(15,4))
sns.scatterplot(data=df_year_market,x="ListingYear",y="count",
                hue="Market"
               ,size="count")
```

```python
#lineplot으로 그리기
plt.figure(figsize=(15,4))
sns.lineplot(data=df_year_market,x="ListingYear",y="count",
                hue="Market")
```

```python
# relplot으로 subplot그리기 - scatter
# subplot을 그릴수있는 sns
sns.relplot(data=df_year_market,x="ListingYear",y="count",col="Market",hue="Market")
```

```python
# relplot으로 subplot그리기 - line
# subplot을 그릴수있는 sns
sns.relplot(data=df_year_market,x="ListingYear",y="count",col="Market",hue="Market",kind="line")
```
